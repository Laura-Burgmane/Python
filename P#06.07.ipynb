ACCESSING COLUMNS IN DATA FRAMES
* 3 WAYS TO ACCESS 1 PARTICULAR COLUMN:
* by name - df.['Gender'] #can access several columns
* by attribute - df.Gender
* by index  - df.iloc[:,;] #[row, column] / [:(means all rows), ;(means all columns)] / example [:, 3]


```python
#ROBERTAS CODE FOR ACCESSING COL:
df['Gender'] # we access column by a name, we can access several columns

df.Gender # accessing column with an attribute

df.iloc[:, 3] # accessing column with index [row, column]
```

```python
df['Gender']#access col by name
```

```python
#how many unique values?
df['Gender'].unique()
```

```python
df.Gender #by attribute; col name can not have spaces in the name; if there are spaces, need to use 'by name' method
```

```python
df.iloc[:, 3] #by index (: means all rows, 3 means 4. column counting from 0)
```

```python
# VALUE COUNTS:
df['Gender'].value_counts()
```

```python
# VALUE COUNTS without dropping empty values(non-number values):
df['Gender'].value_counts(dropna=False)
```

```python
#HOW MANY EMPTY VALUES: to check if all the values ARE empty (true or false)
df.isna()#binary False = 0, True = 1
#.sum() shows how many empty values in total are there in the column
df.isna().sum()
```

```python
#REVERSE OF PREVIOUS: to check if all the values are NOT empty (true or false)
df.notna()
df.notna().all()
```

```python
#drop empty values:
df.dropna()#does not override the existing dataframe, its just an empty method
#to use the new filtered (without empty values) DF, need to:
#either save it in new DF: df_new_dataframe = df.dropna()
#or, with this method to ovveride the current df witht the new one:
#df.dropna(inplace=True)
```

```python
df.dropna(inplace=True) #cleans data from empty values, but does not display them, generates new data frame that can be saved
```

```python
#to save cleaned data: (appears as another file to the left in folder tree to be used in for example excel)
df.to_csv('cleaned_data.csv')
```
