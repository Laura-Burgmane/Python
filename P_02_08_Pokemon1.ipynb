```python
# import required libraries
import pandas as pd #analasys
import matplotlib.pyplot as plt #visualisation
```

```python
pokemon_df = pd.read_csv('/content/Pokemon.csv')
pokemon_df.head(5)
```

```python
# How many rows and columns are in this data?

rows, columns = pokemon_df.shape
print(f'Number of rows is {rows}, number of columns is {columns}')
```

```python
# How many unique Pokemon names are in this table?

unique_names = pokemon_df['Name'].nunique()
unique_names
```

```python
pokemon_df.head(2)
```

```python
# How many pokemons have Attack greater than 100?
attack_greater_than_100 = pokemon_df[pokemon_df['Attack'] > 100]
print(len(attack_greater_than_100))
```

```python
# How many pokemons have Defense greater than 100?
defense_greater_than_100 = pokemon_df[pokemon_df['Defense'] > 100]
print(len(defense_greater_than_100))
```

```python
attack_defense_greater_than_100 = pokemon_df[(pokemon_df['Attack'] > 100) & (pokemon_df['Defense'] > 100)]
print(len(attack_defense_greater_than_100))
```

```python
# VALUE COUNTS!!! Create a grouped DataFrame that counts the number of Pokemons in each group for Type 1.

type1_pokemons = pokemon_df['Type 1'].value_counts().reset_index() #reset index used to avoid column Type1 becoming an index (index must be number)
print(type1_pokemons)
```

```python
 #Bar chart

type1_pokemons.plot(x = 'Type 1', y = 'count', kind = 'bar', color='white', edgecolor='black')
plt.title('Pokemon count by Type 1')
plt.xlabel('Type 1')
plt.ylabel('Count of Pokemons')
```

```python
#another way to visualise for more detailed chart
plt.bar(type1_pokemons['Type 1'], type1_pokemons['count'], color=['thistle','skyblue'], edgecolor='maroon')
plt.xticks(rotation=90)#rotates y axis text vertically
plt.show()#takes out unnecessary text before the chart, remove this and there is page of text before the chart
```

```python
# What is the name of the pokemon that has the biggest Attack value?
pokemon_df['Attack'].max() #method to call the highest value of Attack
pokemon_df['Attack'].idxmax() #method to call the INDEX of the highest value
pokemon_df['Name'].iloc[pokemon_df['Attack'].idxmax()]
highest_attack_name = pokemon_df['Name'].iloc[pokemon_df['Attack'].idxmax()]#iloc = index location
print('Pokemon with the highest Attack rate is called:', highest_attack_name)#printing the pokemon name simple
print(f'Pokemon with the highest Attack rate is called: {highest_attack_name}')#printing the pokemon name with f-string
```

**HISTOGRAM**

```python
#analyse Total for each pokemon
plt.hist(pokemon_df['Total'], color='green', edgecolor = 'black', alpha = 0.7)#alpha for transparency, default 10 bins
plt.xlabel('Total')
plt.ylabel('Number of pokemons (Frequency)')
#adding avg line to the histogram
plt.axvline(pokemon_df['Total'].mean(), linestyle = 'dashed', color = 'grey')
#adding also min and max value
#min
plt.axvline(pokemon_df['Total'].min(), linestyle = 'dashed', color = 'blue')
#max
plt.axvline(pokemon_df['Total'].max(), linestyle = 'dashdot', color = 'orange')
plt.grid(axis = 'x', linestyle = '--', alpha = 0.5)
```

```python
# access counts and bin edges
value = plt.hist(pokemon_df['Total'], bins = 3, color='green', edgecolor='black', alpha=0.7)

hist_values = value[0]
bin_edges = value[1]

print('Histogram counts:', hist_values)
print('Bin edges', bin_edges) #the end value is not included
```
