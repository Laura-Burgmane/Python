```python
##Data visualisation - MATPLOTLIB library
# from synthetic data
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.patches import PathPatch
from matplotlib.path import Path

N = 400
t = np.linspace(0, 2 * np.pi, N)
r = 0.5 + np.cos(t)
x, y = r * np.cos(t), r * np.sin(t)

fig, ax = plt.subplots()
ax.plot(x, y, "k")
ax.set(aspect=1)
```

```python
# Sample data representing the average temperature (in degrees Celsius) from 2010 to 2014
# here is plt library used, more compicated
data = {'Year': [2010, 2011, 2012, 2013, 2014],
        'Avg_Temperature': [14.1, 14.3, 14.5, 14.2, 14.6]}
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame(data)
df
# VERSION ONE

plt.plot(df['Year'], df['Avg_Temperature'], marker = 'o') #first value = x, second value = y
#marker = 'o'
plt.title('Average temperature from 2010 to 2014')#add title
plt.xlabel('Year')#title on x axis
plt.ylabel('Temperature')#title on y axis
plt.xticks(df['Year'])
plt.grid(True)
plt.show()#to visualise the code without any extra text
```

```python
# VERSION TWO

df.plot(x='Year', y='Avg_Temperature', kind='line', marker = 'o')# here is used pandas library

plt.title('Average temperature from 2010 to 2014') #we added title
plt.xlabel('Year') #title on x axis
plt.ylabel('Temperature') #title on y axis
plt.xticks(df['Year']) #year values are rounded
plt.grid(True)
plt.show() #show you the visualization
```

```python
# BAR CHART

plt.bar(df['Year'], df['Avg_Temperature'], color = 'magenta')
plt.title('Average temperature from 2010 to 2014')
plt.xlabel('Year')
plt.ylabel('Temperature')
```

```python
# HORIZONTAL BAR CHART

plt.barh(df['Year'], df['Avg_Temperature'], color = 'green')
plt.title('Average temperature from 2010 to 2014')
plt.xlabel('Avg_Temperature')
plt.ylabel('Year')
```

```python
# HISTOGRAM - presents the distribution of a signle variable, frequency, represents numerical data

data2 = {'Scores': [85, 92, 78, 20, 89, 90, 82, 56, 60, 23, 90, 40, 77]}
df2 = pd.DataFrame(data2)
df2

plt.hist(df2['Scores'], bins = 5, color = 'orange', edgecolor = 'blue')
```

```python
# SCATTER PLOT - used to visualize the relationship between two variables

data3 = {'X': [1,2,3,5,5,2], 'Y':[5,8,7,4,9,5]}

df3 = pd.DataFrame(data3)
df3
plt.scatter(x = df3['X'], y=df3['Y'])
plt.title('Scatter Plot')
plt.grid(True)
```

```python
# PIE CHART - represents parts of a whole

data4 = {'Category': ['A', 'B', 'C', 'D'], 'Values': [15, 30, 25, 40]}
df4 = pd.DataFrame(data4)
df4

plt.pie(df4['Values'], labels = df4['Category'], autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# BOX PLOT - distribution of a dataset and identifies outliers

plt.boxplot(df4['Values'])
plt.grid(True)
plt.show()
```
